Тут я бы хотел дать несколько примеров того, что можно ожидать от нейросетей.
Пример с предсказанием типа ириса (0, 1, 2). Уже встроен в sklearn и вообще это классификатор. 
Здесь собраны данные о длине, ширине чашелистиков и лепестков ирисов и их принадлежности к виду.

Идея в том, что есть массив признаков (X) и есть массив результатов классификации (y), элементы которого принимают количественные значения (классы 0, 1, 2). 

```python
# Импорт модулей в виде методов
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris


iris = load_iris() #Импорт датасета со всеми прибамбасами. Он уже встроен в этот модуль как демо
X = iris.data #Присвоение вектора признаков переменной Х
y = iris.target #Присвоение вектора результатов классификации переменной y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25) #Разбиение всего датасета на тренировочный и тестовый 
dt = DecisionTreeClassifier() #Создание классификатора
dt.fit (X_train, y_train) #Обучение классификатора. Используются только тренировочные датасеты
predicted=dt.predict(X_test) #Классификация по тестовым признакам
```

```python
#Часть массива входных аргументов и результатов для примера
X_view = pd.DataFrame(X)
y_view = pd.DataFrame(y)
X_view, y_view
```

В результате получился массив результатов predicted (это то, что выдала нейросеть). 
То есть, нейросеть взяла вектор аргументов X_train и сравнила его с вектором значений y_train. Так она обучилась.
Нейросеть была уже готовая, обучение заключается только в изменении весов отдельных нейронов (я так думаю). 
У нее там выстроились какие то зависимости внутри. Она решила для себя какие признаки(аргументов) в какой степени влияют на результат классификации (значения). 
ЗАТЕМ!! Нужно было выяснить качество предсказаний. Мы взяли тестовый вектор аргументов X_test и для этого набора аргументов, для каждой строки, попросили классифицировать цветок. Отнести его к одному из видов (0, 1 или 2). 
И она, уже обученная нейросеть, использовала полученные в результате тренировки знания и классифицировала.
y_test - это то, что было заранее задано для X_test.
